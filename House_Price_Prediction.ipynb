importing the dependecies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the House Price Dataset

```python
house_price_dataset= pd.read_csv('/content/BostonHousing.csv')
```

```python
house_price_dataset.head()
```

```python
house_price_dataset.shape
```

```python
house_price_dataset.isnull().sum()
```

```python
house_price_dataset['price']=house_price_dataset['medv']
house_price_dataset.drop(columns='medv', axis=1, inplace=True)
```

```python
house_price_dataset.describe()
```

Understanding the corrolation

```python
# Calculer la matrice de corrlation
correlation_matrix = house_price_dataset.corr()

# Crer une figure pour la heatmap
plt.figure(figsize=(10, 8))

# Afficher la heatmap de la matrice de corrlation
sns.heatmap(correlation_matrix, cbar=True, square=True, fmt='.2f', annot=True, cmap='coolwarm')
```

Spliting the data and the target

```python
X=house_price_dataset.drop(columns='price', axis=1)
Y=house_price_dataset['price']
print(X)
print(Y)
```

Spliting the data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

Model Training

```python
model= XGBRegressor()
model.fit(X_train, Y_train)
```

model evaluation

```python
#accuracy for prediction on training data
X_train_prediction= model.predict(X_train)
print(X_train_prediction)
```

```python
#R squared error
score_1= metrics.r2_score(Y_train, X_train_prediction)
score_2= metrics.mean_absolute_error(Y_train, X_train_prediction)
print(score_1)
print(score_2)
```

```python
sns.scatterplot(y=Y_train, x=X_train_prediction)
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Actual vs Predicted Prices')
plt.show()
```

Prediction on Test Data

```python
test_data_prediction= model.predict(X_test)
score1=metrics.r2_score(Y_test, test_data_prediction)
score2=metrics.mean_absolute_error(Y_test, test_data_prediction)
print(score1)
print(score2)
```

```python
sns.scatterplot(y=Y_test, x=test_data_prediction)
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Actual vs Predicted Prices')
plt.show()
```
